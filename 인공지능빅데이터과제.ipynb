```python
test_data = pd.read_csv('/content/drive/MyDrive/인공지능과 빅데이터 과제/test_data.csv')
test_data
```

```python
test_data.info()
```

```python
X_train = train_data.drop(columns=['target'])
X_train.info()
```

```python
train_data_set = train_data.values
test_data_set = test_data.values
```

```python
X_train = train_data_set[0:952, 0:11]
Y_train = train_data_set[0:952, 11]
X_test = test_data_set[0:238, 0:11]
Y_test = test_data_set[0:238, 11]
```

```python
model = Sequential()
model.add(Dense(units=64, activation='relu', input_dim=11))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, epochs=100, batch_size=10)
model.evaluate(X_test, Y_test)
```
